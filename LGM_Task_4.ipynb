```python
import numpy as np
import pandas as pd
import cv2
from google.colab.patches import cv2_imshow    
from skimage import io
from PIL import Image
import matplotlib.pylab as plt
```

```python
image=cv2.imread("/Flower.jpg")
image1 = cv2.resize(image, (640, 400))
```

```python
cv2_imshow(image1)
```

```python
image_gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
cv2_imshow(image_gray)
```

```python
image_invert = cv2.bitwise_not(image_gray)
cv2_imshow(image_invert)
```

```python
image_smoothing = cv2.GaussianBlur(image_invert, (21, 21),sigmaX=0, sigmaY=0)

cv2_imshow(image_smoothing)
```

```python
def dodgeV2(x, y):
    return cv2.divide(x, 255 - y, scale=256)
```

```python
final_image = dodgeV2(image_gray, image_smoothing)

cv2_imshow(final_image)
```

```python
cv2_imshow(image1)

cv2_imshow(final_image)
```
